# 6. Graficos




```python
# Importar matplotlib ya que Colab ya lo tiene pre-instalado
import matplotlib.pyplot as plt
import numpy as np

#Creacion de array usando Numpy con los datos a usar (Array 1D)
pastel = np.array([Ingreso1, Ingreso2, Ingreso3, Ingreso4])
mylabels = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

#Usar matplotlib para crear grafica
plt.pie(pastel, labels = mylabels, autopct='%1.1f%% ') #autopct = porcentajes
plt.title("Ingresos Totales")
plt.show()
```

```python
#Puntuaciones de las tiendas segun el publico
x = np.array(["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"])
y = np.array([cal_promedio1, cal_promedio2, cal_promedio3, cal_promedio4])


plt.bar(x,y, color = 'gray')
plt.title('Puntuaciones de las tiendas')
plt.show()
```

```python
#Promedio del envio por tienda
#Creamos una array con los valores a usar
yvalores = np.array([envio1,  envio2, envio3, envio4])
xetiquetas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

#graficar
plt.plot(yvalores, marker = 'o')
plt.ylabel('Costo de envio')
plt.xticks(range(len(xetiquetas)), xetiquetas)  # Etiquetas personalizadas en X, con funcion xticks
plt.title('Promedio de envio por tienda')
plt.show()
```

Podemos observar en la gráfica 1 y 3 que la tienda que mayor rendimiento tuvo fue la tienda 1, existe un mayor movimiento de mercancia respecto a las otra tiendas. Pero tambien presenta en promedio un mayor gasto en los envios de su mercancia por lo que seria interesante evaluar si dicho gasto sera sostenible en un periodo a largo plazo y cuanta ganacia podria estar mermando.

Por otro lado, se puede observar que tanto la tienda 2 y 3 gustan mas a los usuarios, lo que tambien se ve reflejado en sus ingresos teniendo el 2do y 3er lugar despues de la tienda 1.

La tienda 4 es la que menores ingresos presenta, pero a su vez es la que menor costos por envio presenta por lo que su rentabilidad se ve beneficiada por lo anterior.

En general, no hay mucha diferenecia entre los ingresos y puntuaciones entre las tiendas. Consiuderop que el factor que determinará que tienda deberá ser removida o cerrada van a ser los costos de envios que dicha tiendas presenten, ya que lo antes mencionado terminará por afectar en la rentabilidad y sostenibilidad del negocio.
